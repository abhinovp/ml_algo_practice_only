```python
## Seaborn 101
## Sources: https://www.youtube.com/watch?v=6GUZXDef2U0
##

import seaborn as sn
import matplotlib.pyplot as plt

## seaborn has default datasets for trails and work.
sn.get_dataset_names()

## Printing our selected data set.
carcra = sn.load_dataset('car_crashes')

print(carcra)
```

```python
## 1.1 Dist plot - univariate

sn.displot(carcra['not_distracted'],kde='true') ##kernel density estimation

sn.displot(carcra['not_distracted'],kde='true',bins = 25) ##bins/buckets -> number of sectors needed.
```

```python
## 1.1 Joint plot - used to compare 2 distribution

sn.jointplot(x='speeding',y='alcohol',data = carcra,kind='reg')

sn.jointplot(x='speeding',y='alcohol',data = carcra,kind='kde') ##KDE is cooler

sn.jointplot(x='speeding',y='alcohol',data = carcra,kind='hex') ##KDE is cooler
```

```python
## 1.3 KDE plots

sn.kdeplot(carcra['alcohol']) ## Regular KDE
```

```python
## 2 Pair plot  - compares data sets in pairs

sn.pairplot(carcra) ## alongside hue = a column and palatte="blues" can be added.

```

```python
## 3 . Rug plots ( carpet end thread kinda stuff)

sn.rugplot(carcra['speeding'],height = 0.1)
```

```python
## 4. Styling all plots in graph.

sn.set_style("whitegrid") # ("dark")  ("ticks") ("darkgrid") ("whitegrid")

sn.jointplot(x='speeding',y='alcohol',data = carcra,kind='reg')


## Setting context( font and sizes )

sn.set_context('talk',font_scale = 1.1) ## Default is 'paper'

sn.jointplot(x='speeding',y='alcohol',data = carcra,kind='reg')

## Removing axis lines

sn.despine(left="True")




```

```python
## 5.1 Categorical plot

## Bar plot
plt.figure(figsize=(8,5))

sn.barplot(data=carcra,x="speeding",y="alcohol")


#sn.set_context('talk', font_scale = 1)
```

```python
## 5.2 Count plot

#sn.get_dataset_names() 

tipsd = sn.load_dataset('tips')

sn.countplot(x="sex",data = tipsd) #,x="speeding",y="alcohol")
```

```python
## 5.3 Box plot

sn.boxplot(x="day",y="total_bill",data=tipsd,hue="sex")
```

```python
## 5.4 Violin plot =  Box plot + KDE

sn.violinplot(x="day",y="total_bill",data=tipsd,hue="sex")

print( "Onto another plot")

```

```python
sn.violinplot(x="day",y="total_bill",data=tipsd,hue="sex",split = "true")
```

```python
## 5.4 Strip plot 

sn.stripplot(x="day",y="total_bill",data=tipsd,hue="sex",dodge = "true")
```

```python
## 5.5 Swarm plot = like a strip plot with dots are spread for clarity

sn.swarmplot(x="day",y="total_bill",data=tipsd,hue="sex")
```

```python
## Swarm with Violin plot
sn.violinplot(x="day",y="total_bill",data=tipsd)
sn.swarmplot(x="day",y="total_bill",data=tipsd,color="white")
```

```python
### Pallates 
### style and context
### colors and clarity

plt.figure(figsize = (8,5))

sn.set_style("darkgrid")
sn.set_context("talk")
## using a box plot
sn.boxplot(x="day",y="total_bill",data=tipsd,hue="sex",palette="magma")
plt.legend(loc='upper right') #(loc=1) ## 1,2,3,4 order starts from top right and ends in bottom right
```
