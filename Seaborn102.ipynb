```python
### Seaborn 102
## Sources: https://www.youtube.com/watch?v=6GUZXDef2U0
##

import seaborn as sn
import matplotlib.pyplot as plt

## seaborn has default datasets for trails and work.
sn.get_dataset_names()

## Printing our selected data set.
tipsy = sn.load_dataset('tips')

print(tipsy)
```

```python
## 1 Matrix plots 
## 1.1 Heat map

plt.figure(figsize=(8,4))

sn.set_context('paper',font_scale = 1.2)

tipscorr = tipsy.corr()

sn.heatmap(tipscorr,annot=True,cmap ="Blues") ## Heatmap of correleation
```

```python
## 1.2 Pivot - plot : heatmap

flights =sn.load_dataset('flights')

fpivot = flights.pivot_table(index="month",values="passengers",columns="year")

fpivot ##generated pivot table

sn.heatmap(fpivot,cmap="Blues")
```

```python
fpivot ##generated pivot table

sn.heatmap(fpivot,cmap="Blues",linewidth=1)
```

```python
## 1.3 Cluster maps - Basically hierarachically clustered heatmap.<skipped>
## 1.4 Pair grid - just to pick 2 plots grid side by side

tips_g = sn.PairGrid(tipsy,hue="day")
tips_g.map(plt.scatter)
```

```python
## Can also split for upper,diag and lower

tips_g = sn.PairGrid(tipsy,hue="day")

tips_g.map_upper(plt.scatter)
tips_g.map_diag(plt.hist)
tips_g.map_lower(sn.kdeplot)
```

```python
## 1.5 Facet grid - Similar grid with deeper criss cross of value plots

tips_f = sn.FacetGrid(tipsy,col="time",row="smoker")

tips_f.map(plt.scatter,'total_bill',"tip")
```

```python
## 1.5 Facet grid - Similar grid with deeper criss cross of value plots

tips_f = sn.FacetGrid(tipsy,col="time",row="smoker",height=3,aspect=1.2)

tips_f.map(plt.scatter,'total_bill',"tip")
```

```python
## 1.5 Facet grid - Similar grid with deeper criss cross of value plots

tips_f = sn.FacetGrid(tipsy,col="time",row="smoker",height=3,aspect=1.2,col_order=["Dinner","Lunch"])

tips_f.map(plt.scatter,'total_bill',"tip")
```

```python
## 1.6 Regression plots

plt.figure(figsize=(38,24))

sn.set_context('talk',font_scale = 1.2)

sn.lmplot(data=tipsy,x="tip",y="total_bill",hue="sex",markers=[".","^"],height=8,aspect=1.2)
```
