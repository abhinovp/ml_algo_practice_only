```python
# Decision Tree & Gini Impurity
# Source: https://www.youtube.com/watch?v=PHxYNGo8NcI&list=PLeo1K3hjS3uvCeTYTeyfe0-rN5r8zn9rw&index=11
# data files stored in data_files
# Excercise: To predict survival rate


import pandas as ps
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as ny


tdf = ps.read_csv(r'C:/Users/140115/PycharmProjects/JupyterMoons/data_files/titanic.csv')
tdf.head()
```

```python
target= tdf['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
tdf['gender_n'] = le_gender.fit_transform(tdf['Sex'])
tdf.head()
```

```python
# Tapping the input set from the main DF
# Prepare data(Clean, fill and construct)
inputset = tdf.drop(['Sex','PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')

#Filling missing values with mean
inputset['Age'].fillna(inputset['Age'].mean(),inplace=True)
inputset['Fare'].replace(0,inputset['Fare'].mean(),inplace=True)


#Re-checking new DF
inputset.loc[170:183,:]
```

```python
#Seperating target

target_n = inputset['Survived']
target_n

in2 = inputset.drop('Survived',axis='columns')

#T-T-S
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(in2,target_n,test_size=0.2)

X_test
```

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier()
tree_model.fit(X_train,y_train)

```

```python
#predict and score
tree_model.predict(X_test)

tree_model.score(X_test,y_test)
```
